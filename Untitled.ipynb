```python
import csv

ds_nhan_vien = [
    {"ma": "NV001", "ten": "Nguyen Van A", "phong_ban": "IT", "vi_tri": "Developer", "sdt": "0901234567"},
    {"ma": "NV002", "ten": "Tran Thi B", "phong_ban": "HR", "vi_tri": "Recruiter", "sdt": "0912345678"},
    {"ma": "NV003", "ten": "Le Van C", "phong_ban": "Sales", "vi_tri": "Manager", "sdt": "0923456789"},
]


# Thêm nhân viên
def them_nhan_vien():
    ma = input("Nhập mã nhân viên: ")
    ten = input("Nhập họ tên: ")
    phong_ban = input("Nhập phòng ban: ")
    vi_tri = input("Nhập vị trí: ")
    sdt = input("Nhập số điện thoại: ")

    nv = {
        "ma": ma,
        "ten": ten,
        "phong_ban": phong_ban,
        "vi_tri": vi_tri,
        "sdt": sdt
    }

    ds_nhan_vien.append(nv)
    print("Thêm nhân viên thành công!\n")


# Hiển thị danh sách
def hien_thi():
    if len(ds_nhan_vien) == 0:
        print("Danh sách rỗng!\n")
        return

    print("\n===== DANH SÁCH NHÂN VIÊN =====")
    print(f"{'Mã NV':<10} | {'Họ và Tên':<25} | {'Phòng Ban':<15} | {'Vị Trí':<15} | {'Số Điện Thoại':<15}")
    print("-" * 88)

    for nv in ds_nhan_vien:
        print(f"{nv['ma']:<10} | {nv['ten']:<25} | {nv['phong_ban']:<15} | {nv['vi_tri']:<15} | {nv['sdt']:<15}")

    print("=" * 88 + "\n")


# Sắp xếp theo tên
def sap_xep_ten():
    ds_nhan_vien.sort(key=lambda x: x["ten"])
    print("Đã sắp xếp theo tên!\n")


# Sắp xếp theo phòng ban
def sap_xep_phong_ban():
    ds_nhan_vien.sort(key=lambda x: x["phong_ban"])
    print("Đã sắp xếp theo phòng ban!\n")


# Tìm theo mã
def tim_theo_ma():
    ma = input("Nhập mã nhân viên cần tìm: ")

    for nv in ds_nhan_vien:
        if nv["ma"].lower() == ma.lower():
            print("\n===== KẾT QUẢ TÌM KIẾM =====")
            print(f"{'Mã NV':<10} | {'Họ và Tên':<25} | {'Phòng Ban':<15} | {'Vị Trí':<15} | {'Số Điện Thoại':<15}")
            print("-" * 88)
            print(f"{nv['ma']:<10} | {nv['ten']:<25} | {nv['phong_ban']:<15} | {nv['vi_tri']:<15} | {nv['sdt']:<15}")
            print("=" * 88 + "\n")
            return

    print("Không tìm thấy!\n")


# Tìm theo tên
def tim_theo_ten():
    ten = input("Nhập tên nhân viên cần tìm: ")
    ket_qua = [nv for nv in ds_nhan_vien if ten.lower() in nv["ten"].lower()]

    if len(ket_qua) == 0:
        print("Không tìm thấy!\n")
        return

    print("\n===== KẾT QUẢ TÌM KIẾM =====")
    print(f"{'Mã NV':<10} | {'Họ và Tên':<25} | {'Phòng Ban':<15} | {'Vị Trí':<15} | {'Số Điện Thoại':<15}")
    print("-" * 88)

    for nv in ket_qua:
        print(f"{nv['ma']:<10} | {nv['ten']:<25} | {nv['phong_ban']:<15} | {nv['vi_tri']:<15} | {nv['sdt']:<15}")

    print("=" * 88 + "\n")


# Xóa nhân viên
def xoa_nhan_vien():
    ma = input("Nhập mã nhân viên cần xóa: ")

    for nv in ds_nhan_vien:
        if nv["ma"].lower() == ma.lower():
            ds_nhan_vien.remove(nv)
            print("Xóa thành công!\n")
            return

    print("Không tìm thấy nhân viên!\n")


# Lưu file CSV
def luu_csv():
    with open("nhanvien.csv", "w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)

        writer.writerow(["MaNV", "HoTen", "PhongBan", "ViTri", "SoDienThoai"])

        for nv in ds_nhan_vien:
            writer.writerow([
                nv["ma"],
                nv["ten"],
                nv["phong_ban"],
                nv["vi_tri"],
                nv["sdt"]
            ])

    print("Đã lưu file CSV!\n")


# Đọc file CSV
def doc_csv():
    try:
        with open("nhanvien.csv", "r", encoding="utf-8") as file:
            reader = csv.reader(file)

            next(reader)

            ds_nhan_vien.clear()

            for row in reader:
                nv = {
                    "ma": row[0],
                    "ten": row[1],
                    "phong_ban": row[2],
                    "vi_tri": row[3],
                    "sdt": row[4] if len(row) > 4 else ""
                }

                ds_nhan_vien.append(nv)

        print("Đọc file thành công!\n")

    except FileNotFoundError:
        print("Chưa có file CSV!\n")


# Menu chính
while True:

    print("\n===== QUẢN LÝ HỒ SƠ NHÂN VIÊN =====")
    print("1. Thêm nhân viên")
    print("2. Hiển thị danh sách")
    print("3. Sắp xếp theo tên")
    print("4. Sắp xếp theo phòng ban")
    print("5. Tìm theo mã")
    print("6. Tìm theo tên")
    print("7. Xóa nhân viên")
    print("8. Lưu CSV")
    print("9. Đọc CSV")
    print("0. Thoát")

    chon = input("Nhập lựa chọn: ")

    if chon == "1":
        them_nhan_vien()

    elif chon == "2":
        hien_thi()

    elif chon == "3":
        sap_xep_ten()

    elif chon == "4":
        sap_xep_phong_ban()

    elif chon == "5":
        tim_theo_ma()

    elif chon == "6":
        tim_theo_ten()

    elif chon == "7":
        xoa_nhan_vien()

    elif chon == "8":
        luu_csv()

    elif chon == "9":
        doc_csv()

    elif chon == "0":
        print("Kết thúc chương trình!")
        break

    else:
        print("Lựa chọn không hợp lệ!")
```
